````python
# üßÆ Beginner Python Mini Projects

These are simple yet fun Python projects to practice your basics:
1. **Simple Calculator**
2. **Number Guessing Game**
3. **To-Do List Manager**

---

## ‚öôÔ∏è 1. Simple Calculator

Let‚Äôs create a calculator that can perform **addition, subtraction, multiplication, and division**.

```python
# Simple Calculator

def calculator():
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    choice = input("Enter choice (1/2/3/4): ")

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == '1':
        print(f"The result is: {num1 + num2}")
    elif choice == '2':
        print(f"The result is: {num1 - num2}")
    elif choice == '3':
        print(f"The result is: {num1 * num2}")
    elif choice == '4':
        if num2 != 0:
            print(f"The result is: {num1 / num2}")
        else:
            print("Error: Cannot divide by zero.")
    else:
        print("Invalid input!")

calculator()
‚úÖ Try it: Run and choose an operation (1‚Äì4).

üéØ 2. Number Guessing Game
The computer picks a random number, and you try to guess it!


import random

def guessing_game():
    number_to_guess = random.randint(1, 20)
    attempts = 0

    print("I'm thinking of a number between 1 and 20.")

    while True:
        guess = int(input("Take a guess: "))
        attempts += 1

        if guess < number_to_guess:
            print("Too low! Try again.")
        elif guess > number_to_guess:
            print("Too high! Try again.")
        else:
            print(f"üéâ Congratulations! You guessed it in {attempts} tries.")
            break

guessing_game()
‚úÖ Try it: See how many attempts it takes you to win!

üìù 3. To-Do List Manager
A simple text-based task manager that lets you add, view, and remove tasks.


todo_list = []

def show_menu():
    print("\n--- To-Do List Menu ---")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

def add_task():
    task = input("Enter a new task: ")
    todo_list.append(task)
    print(f"‚úÖ Task '{task}' added!")

def view_tasks():
    if not todo_list:
        print("No tasks yet!")
    else:
        print("\nYour Tasks:")
        for i, task in enumerate(todo_list, start=1):
            print(f"{i}. {task}")

def remove_task():
    view_tasks()
    try:
        task_num = int(input("Enter task number to remove: "))
        removed = todo_list.pop(task_num - 1)
        print(f"‚ùå Task '{removed}' removed.")
    except (IndexError, ValueError):
        print("Invalid input. Please try again.")

def main():
    while True:
        show_menu()
        choice = input("Choose an option (1-4): ")

        if choice == '1':
            add_task()
        elif choice == '2':
            view_tasks()
        elif choice == '3':
            remove_task()
        elif choice == '4':
            print("Goodbye! üëã")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

main()
‚úÖ Try it: Add, view, and remove your daily tasks!

````
