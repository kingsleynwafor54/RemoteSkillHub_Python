```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6), dpi=100)
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title("Basic Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(10, 4))

# Left graph
plt.subplot(1, 2, 1)   # 1 row, 2 columns, 1st plot
plt.plot([1, 2, 3], [2, 4, 6], color='blue')
plt.title("Plot 1: Line")

# Right graph
plt.subplot(1, 2, 2)   # 1 row, 2 columns, 2nd plot
plt.bar([1, 2, 3], [2, 4, 6], color='orange')
plt.title("Plot 2: Bar")

# Show the entire figure
plt.tight_layout()  # Prevent overlap
plt.show()
```

```python
plt.figure(figsize=(2, 4))

plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [3, 6, 9])
plt.title("1")

plt.subplot(2, 2, 2)
plt.scatter([1, 2, 3], [3, 2, 5])
plt.title("2")

plt.subplot(2, 2, 3)
plt.bar([1, 2, 3], [5, 7, 3])
plt.title("3")

plt.subplot(2, 2, 4)
plt.hist([1, 2, 1, 3, 2, 1])
plt.title("4")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Top-left
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [2, 4, 6], color='blue')
plt.title("Line Plot")

# Top-right
plt.subplot(2, 2, 2)
plt.bar([1, 2, 3], [3, 1, 4], color='orange')
plt.title("Bar Chart")

# Bottom-left
plt.subplot(2, 2, 3)
plt.scatter([1, 2, 3], [3, 2, 5], color='green')
plt.title("Scatter Plot")

# Bottom-right
plt.subplot(2, 2, 4)
plt.hist([1, 2, 1, 3, 2, 1], color='purple')
plt.title("Histogram")

plt.tight_layout()
plt.show()
```
