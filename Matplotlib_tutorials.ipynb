```python
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(10, 4))

# Left graph
plt.subplot(1, 2, 1)   # 1 row, 2 columns, 1st plot
plt.plot([1, 2, 3], [2, 4, 6], color='blue')
plt.title("Plot 1: Line")

# Right graph
plt.subplot(1, 2, 2)   # 1 row, 2 columns, 2nd plot
plt.bar([1, 2, 3], [2, 4, 6], color='orange')
plt.title("Plot 2: Bar")

# Show the entire figure
plt.tight_layout()  # Prevent overlap
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Top-left
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [2, 4, 6], color='blue')
plt.title("Line Plot")

# Top-right
plt.subplot(2, 2, 2)
plt.bar([1, 2, 3], [3, 1, 4], color='orange')
plt.title("Bar Chart")

# Bottom-left
plt.subplot(2, 2, 3)
plt.scatter([1, 2, 3], [3, 2, 5], color='green')
plt.title("Scatter Plot")

# Bottom-right
plt.subplot(2, 2, 4)
plt.hist([1, 2, 1, 3, 2, 1], color='purple')
plt.title("Histogram")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 5))

plt.plot([1, 2, 3, 4], [10, 20, 25, 30], label="Sales 2024", color='blue')
plt.plot([1, 2, 3, 4], [15, 18, 22, 28], label="Sales 2025", color='orange')
plt.plot([1, 2, 3, 4], [12, 14, 25, 27], label="Sales 2025", color='red')

plt.title("Sales Growth Comparison")
plt.xlabel("Quarter")
plt.ylabel("Revenue (in $1000s)")

#  This line adds the legend box
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2 * np.pi, 100)  # 100 evenly spaced values between 0 and 2
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.figure(figsize=(10, 6))  # canvas size

# Sine
plt.plot(x, y_sin, label='sin(x)', color='blue')

# Cosine
plt.plot(x, y_cos, label='cos(x)', color='red')

# Tangent
plt.plot(x, y_tan, label='tan(x)', color='green')

# Add title and axis labels
plt.title("Trigonometric Functions")
plt.xlabel("Angle (radians)")
plt.ylabel("Function value")

# Add grid
plt.grid(True, linestyle='--', alpha=0.5)

# Add legend
plt.legend()

# Display the graph
plt.show()
```
